```python
import gensim.downloader as api
```

```python
# Carregar embeddings Word2Vec pré-treinados
w2v_model = api.load("word2vec-google-news-300")

# Obter vetor de palavra para uma palavra específica
word_vector = w2v_model['futebol']
print("Vetor Word2Vec para a palavra 'futebol':", word_vector)

# Encontrar palavras mais similares
similar_words = w2v_model.most_similar('futebol', topn=5)
print("Palavras mais similares a 'futebol':", similar_words)
```
